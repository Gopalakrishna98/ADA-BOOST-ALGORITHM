```python
import numpy as np
import pandas as pd
#from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.ensemble import AdaBoostClassifier
import seaborn as sns 
import pandas as pd
```

```python
data=sns.load_dataset('iris')
data.head(3)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=AdaBoostClassifier(n_estimators=4,learning_rate=0.5)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
metrics.accuracy_score(y_train,model.predict(x_train))
```

```python
metrics.confusion_matrix(y_train,model.predict(x_train))
```

## USING FUN DATA

```python
data=pd.read_csv('fun data.csv')
data.head(3)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
for i in np.linspace(0.01,1,40):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)    
    model=AdaBoostClassifier(n_estimators=500,learning_rate=i)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(metrics.accuracy_score(y_test,y_pred))
```

```python
y_test_pred=model.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred)
```

```python
metrics.accuracy_score(y_train,model.predict(x_train))
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
```

```python
for i in np.linspace(0.01,1,10):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)    
    model=AdaBoostClassifier(n_estimators=50,base_estimator=l,learning_rate=i)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(metrics.accuracy_score(y_test,y_pred))
```

##  ADA BOOST FOR DIABETES REGRESSOR DATA

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
df=pd.read_csv('diabetes_reg_data.csv')
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=AdaBoostRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn import metrics
metrics.mean_squared_error(y_test,y_pred)
```

```python
metrics.mean_absolute_error(y_test,y_pred)
```
